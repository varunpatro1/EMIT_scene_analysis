```python
n, bins, patches = plt.hist(bad_df['zen'], bins = 8)
plt.title('Zenith Angle Histogram of Bad Scenes')
plt.xlabel('Zenith (deg)')
```

```python
n, bins, patches = plt.hist(bad_df['slope'], bins = 8)
plt.title('Slope Histogram of Bad Scenes')
plt.xlabel('Slope')
```

```python
n, bins, patches = plt.hist(bad_df['wv range'], bins = 8)
plt.title('WV Range Histogram of Bad Scenes')
plt.xlabel('WV Range (g/cm^2)')
```

```python
n, bins, patches = plt.hist(bad_df['wv med'], bins = 8)
plt.title('WV Median Histogram of Bad Scenes')
plt.xlabel('WV Median (g/cm^2)')
```

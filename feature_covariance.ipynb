```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pickle
```

```python
df = pd.read_pickle('../../emit_data')
df = df.iloc[:,:8]
df.head()
```

```python
corr_matrix = df.corr()
corr_matrix
```

```python
from pandas.plotting import scatter_matrix
attributes = ['elev med', 'zen', 'slope', 'wv range', 'wv med', 'total cloud fraction', 'lat', 'lon']
scatter_matrix(df[attributes], figsize = (12, 8), alpha = 0.3);
```

```python
plt.scatter(df['wv med'], df['elev med'], alpha = 0.1)
plt.title('Elevation Median vs WV Median')
plt.xlabel('WV Median (g/cm^2)')
plt.ylabel('Elevation Median (km)')
```

```python
plt.scatter(df['wv range'], df['elev med'], alpha = 0.1)
plt.title('Elevation Median vs WV Range')
plt.xlabel('WV Range (g/cm^2)')
plt.ylabel('Elevation Median (km)')
```

```python
attributes = ['elev med', 'zen', 'slope', 'wv med', 'total cloud fraction', 'wv range']
scatter_matrix(df[attributes], figsize = (12, 8), alpha = 0.3);
```

```python
plt.scatter(df['elev med'], df['wv range'], alpha = 0.25)
plt.title('WV Range vs Elev Med')
plt.ylabel('WV Range (g/cm^2)')
plt.xlabel('Elevation Median (km)')
```

```python
plt.scatter(df['zen'], df['wv range'], alpha = 0.25)
plt.title('WV Range vs Elev Med')
plt.ylabel('WV Range (g/cm^2)')
plt.xlabel('Solar Zenith Angle')
```

```python
from pandas.plotting import scatter_matrix
attributes = ['elev med', 'zen', 'slope', 'wv range', 'wv med']
scatter_matrix(df[attributes], figsize = (12, 8), alpha = 0.3, grid = True);
```
